Repeat(RANDOM SEED & RECORD)

```python
import subprocess
import yaml
import numpy as np
from tqdm import tqdm

num_runs = 10

progress_bar = tqdm(total=num_runs, desc='Running')
for _ in range(num_runs):
    seed = np.random.randint(0, 100)
    with open('ODGS/config/config-optuna.yaml', 'r') as file:
        config = yaml.safe_load(file)
        config['seed'] = seed
    with open('ODGS/config/config-optuna.yaml', 'w') as file:
        yaml.dump(config, file)
    print("Current seed:", seed)
    subprocess.run(["python", "MDD/src/train.py"])

    # 更新进度条
    progress_bar.update(1)

# 关闭进度条
progress_bar.close()
```
